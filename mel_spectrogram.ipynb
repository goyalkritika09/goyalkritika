```python
import librosa
import librosa.display
import glob2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

dir_name=r"E:\Acoustic\Dataset"     #AUDIO FILES ARE PRSENT HERE

label=0		#CLASSIFICATION LABEL
f_len=2048	#FRAME LENGTH
h_len=1024	#HOP LENGTH

features=[]
frames=[]
concatinatelis=[]
allfreatures=[]
number=0
#save_path = 'test.jpg'
for file in glob2.glob(dir_name+'\\*.wav'):
    y,sr=librosa.load(file)
    filename, file_extension = os.path.splitext(file)
    print(filename)
    print(file_extension)    
    save_path = filename+'.jpg'
    melspectrogram = librosa.feature.melspectrogram(y, sr=sr, n_fft=2048,hop_length=h_len,n_mels=128)
    melspectrogram_DB = librosa.power_to_db(melspectrogram, ref=np.max)
    librosa.display.specshow(melspectrogram_DB, sr=sr, hop_length=h_len);
    #plt.colorbar(format='%+2.0f dB'); 
    plt.savefig(save_path, bbox_inches='tight', pad_inches=-0.1)
    plt.show()
    plt.close()



    

    
    

   
    




```
