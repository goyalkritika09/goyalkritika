```python
import librosa
import glob2
import numpy as np
import pandas as pd

dir_name=r"E:\Acoustic\Dataset\EMODB\DISGUST"     #AUDIO FILES ARE PRSENT HERE

label=0		#CLASSIFICATION LABEL
f_len=2048	#FRAME LENGTH
h_len=1024	#HOP LENGTH

features=[]
frames=[]
concatinatelis=[]
allfreatures=[]
number=0
for file in glob2.glob(dir_name+'\\*.wav'):
    y,sr=librosa.load(file)
    mfccs = librosa.feature.mfcc(y, sr, n_mfcc=41, n_fft=f_len, hop_length=h_len)
    #print(mfccs.shape)
    allfreaturesrows=pd.DataFrame(mfccs) #TO PRINT ALL FRAMES INDIVIDUALLY
    #print(allfreaturesrows)
    allfreatures.append(allfreaturesrows.transpose()) #append all frames in list
    df_final = pd.concat(allfreatures) #concatinate all frames of all to make one file
    df_final['labels']='Disgust'
    df_final.to_csv(r'E:\Acoustic\Dataset\Experiment2withalframes\Disgustallframe.csv')
    
```
